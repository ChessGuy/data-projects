## **4. Data Preprocessing** <a class="anchor" id="data-preprocess"></a>

```python
names = board_game_data['Name']
features = board_game_data.drop('Name', axis=1)
```

### 4.1 Custom Transformer <a class="anchor" id="custom"></a>

There are several columns that need to be removed before feeding the data into the model, and a column transformer will be used to do this. 

The following columns will be removed:

  * ID:  Not needed and is missing values
  * Year Published:  The model's job is to recommend games that are similar, and having this field may skew the model towards games within the time period of the input, which is undesirable.
  * BGG Rank:  Keeping this will skew the model to similar ranking games.

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import make_pipeline

ID_index = features.columns.get_loc('ID')
year_index = features.columns.get_loc('Year Published')
rank_index = features.columns.get_loc('BGG Rank')

class ColumnRemovalTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        self = self
        
    def fit(self, X, y = None):
        return self

    def transform(self, X):
        X = np.delete(X, [ID_index, year_index, rank_index], 1)
        return np.c_[X]
    
col_remover = make_pipeline(
    ColumnRemovalTransformer()
)
```

```python
# Test
ColumnRemovalTransformer().fit_transform(features).shape
```

### 4.2 Numerical Pipeline <a class="anchor" id="numeric"></a>

```python
num_attributes = features.select_dtypes(np.number).drop(['ID', 'Year Published', 'BGG Rank'], axis=1).columns.tolist()
num_attributes
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline

num_pipeline = make_pipeline(
    SimpleImputer(strategy='mean'),
    StandardScaler()
)
```

```python
# Test
num_pipeline.fit_transform(features[num_attributes])
```

### 4.3 Categorical Pipeline <a class="anchor" id="category"></a>

```python
cat_attributes = ['Domains', 'Mechanics']
cat_attributes
```

```python
def string_split(data):
    df = pd.DataFrame(data)
    domains_df = df['Domains'].str.split(',', expand=True)
    mechanics_df = df['Mechanics'].str.split(',',expand=True)
    return pd.concat([domains_df, mechanics_df], axis=1)
```

```python
string_split(features[cat_attributes])
```

```python
from sklearn.preprocessing import MultiLabelBinarizer, FunctionTransformer, OneHotEncoder

cat_pipeline = make_pipeline(
    SimpleImputer(strategy='constant', fill_value='None').set_output(transform="pandas"),
    FunctionTransformer(func=string_split, validate=False).set_output(transform="pandas"),
    OneHotEncoder(sparse_output=False)
)
```

```python
# Test
pd.DataFrame(cat_pipeline.fit_transform(features[cat_attributes]))
```

### 4.4 Column Transformer <a class="anchor" id="column"></a>

```python
from sklearn.compose import ColumnTransformer
from sklearn.compose import make_column_selector

preprocessor = ColumnTransformer([
    ('num', num_pipeline, num_attributes),
    ('cat', cat_pipeline, cat_attributes)
], remainder='drop')
```

```python
# Test
preprocessor.fit_transform(features).shape
```

```python
features_fitted = preprocessor.fit_transform(features)
pd.DataFrame(features_fitted)
```

## **5. Model Selection and Training** <a class="anchor" id="selection"></a>

### 5.1 Clustering Models <a class="anchor" id="cluster"></a>

```python
%%time
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2000, random_state=0, n_init='auto')
kmeans.fit_predict(features_fitted)
```

```python
pd.DataFrame(kmeans.labels_).value_counts()
```

```python
%%time
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(min_samples=3)
dbscan.fit_predict(features_fitted)
```

```python
pd.DataFrame(dbscan.labels_).value_counts()
```

```python
%%time
from sklearn.cluster import AgglomerativeClustering

agg_model = AgglomerativeClustering(n_clusters=2000, linkage='ward')
agg_model.fit_predict(features_fitted)
```

```python
pd.DataFrame(agg_model.labels_).value_counts()
```

```python
%%time
from sklearn.cluster import HDBSCAN

hdbscan = HDBSCAN(min_cluster_size=10)
hdbscan.fit_predict(features_fitted)
```

```python
pd.DataFrame(hdbscan.labels_).value_counts()
```

```python
from k_means_constrained import KMeansConstrained

kmeans_contstrained = KMeansConstrained(size_min=5, size_max=15,random_state=0)
kmeans_contstrained.fit_predict(features_fitted)
```

The description of the columns from the dataset author's Kaggle post are listed below:

|variable                 |class     |description |
|:---|:---|:-------|
|ID                       |int       | BoardGameGeek ID Number |
|Name                     |character | Board game name  |
|Year Published           |int       | Year published  |
|Min Players              |int       | The minimum suggested number of players to play the game |
|Max Players              |int       | The maximum suggested number of players to play the game |
|Play Time                |int       | Average play time in minutes as suggested by the game creators |
|Min Age                  |int       | Age rating |
|Users Rated              |int       | Number of users who rated the game |
|Rating Average           |float64   | Average of user ratings |
|BGG Rank                 |int       | BoardGameGeek ranking |
|Complexity Average       |float64   | Average of user ratings for complexity from 1 - 5 |
|Owned Users              |int       | Number of users who own the game |
|Mechanics                |character | List of game mechanics for that game |
|Domains                  |character | List of game subgenres |
